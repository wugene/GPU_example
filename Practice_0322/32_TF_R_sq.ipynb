```python
import numpy as np
import tensorflow as tf
```

```python
a = np.random.randint(0, 101, size=(1000,100))
b = np.random.randint(0, 101, size=(1,100))
```

```python
x_tf = tf.placeholder('float32', (None,None))
y_tf = tf.placeholder('float32', (1,None))
```

```python
x바 = tf.reduce_mean(x_tf, axis=1, keep_dims=True)
y바 = tf.reduce_mean(y_tf)
```

```python
x프라임 = x_tf - x바
y프라임 = y_tf - y바
```

```python
x_SS = tf.reduce_sum((x프라임 * x프라임), axis=1, keep_dims=True)
y_SS = tf.reduce_sum( y프라임 * y프라임)
```

```python
x프라임y프라임 = tf.matmul(x프라임, y프라임, transpose_b=True)
```

```python
r_tf = x프라임y프라임 / tf.sqrt(x_SS * y_SS)
```

```python
with tf.Session() as sess:
    res = sess.run(r_tf, feed_dict={x_tf: a, y_tf: b})
```
