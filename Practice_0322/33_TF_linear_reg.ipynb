```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
x_data = np.random.normal(0., 1., size=100)
y_data = [ x * (-0.3) + 0.7 + np.random.normal(0., .1) for x in x_data ]
```

```python
plt.plot(x_data, y_data, 'bo')
plt.show()
```

```python
W = tf.Variable(tf.zeros([1]))
b = tf.Variable(tf.zeros([1]))
y = W * x_data + b
```

```python
loss = tf.reduce_mean(tf.square(y - y_data))
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.2)
train = optimizer.minimize(loss)
```

```python
init = tf.global_variables_initializer()

sess = tf.Session()
sess.run(init)
```

```python
history = dict()
for step in range(10):
    sess.run(train)
    history[step] = sess.run(W), sess.run(b)
```

```python
plt.plot(x_data, y_data, 'bo')
for step in range(10):
    plt.plot(x_data, history[step][0] * x_data + history[step][1])
plt.show()
```

```python
optimizer = tf.train.AdamOptimizer()
train = optimizer.minimize(loss)
```

```python
init = tf.global_variables_initializer()

sess = tf.Session()
sess.run(init)
```

```python
history = dict()
for step in range(10):
    sess.run(train)
    history[step] = sess.run(W), sess.run(b)
```

```python
plt.plot(x_data, y_data, 'bo')
for step in range(10):
    plt.plot(x_data, history[step][0] * x_data + history[step][1])
plt.show()
```
