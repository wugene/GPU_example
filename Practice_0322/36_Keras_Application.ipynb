```python
import numpy as np

from keras.preprocessing import image
from keras.applications.inception_v3 import InceptionV3, \
    preprocess_input, decode_predictions
```

## email : wugeneyang@gmail

```python
img = image.load_img('a_dog.jpg', target_size=(299,299))
```

```python
img
```

```python
x = image.img_to_array(img)
```

```python
x.shape
```

```python
X = np.expand_dims(x, axis=0)
```

```python
X.shape
```

```python
X = preprocess_input(X)
```

```python
model = InceptionV3()
```

```python
model.summary()
```

```python
y_pred = model.predict(X)
```

```python
y_pred.argmax()
```

```python
top3 = decode_predictions(y_pred, top=3)
```
